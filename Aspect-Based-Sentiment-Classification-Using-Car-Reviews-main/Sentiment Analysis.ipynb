```python
import pandas as pd
#reading the data
data = pd.read_csv("Review1.csv")
data.head()
```

```python
pd.set_option('display.max_colwidth', -1)
original = data
data= data [['Rating','Review']]
data.head()
```

```python
#checking the count of the dependent variable
data['Rating'].value_counts()
```

```python
#library that contains punctuation
import string
string.punctuation
```

```python
count = 0
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    #global count
    #count=count+1
    #print(count," ",text,"\n")
    return punctuationfree
#storing the puntuation free text
data['clean_msg']= data['Review'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
data['msg_lower']= data['clean_msg'].apply(lambda x: x.lower())
```

```python
import nltk
nltk.download('punkt')
import re
def tokenization(text):
    tokens = nltk.word_tokenize(text)
    #tokens = re.split('W+',text)
    return tokens
#applying function to the column
data['msg_tokenied']= data['msg_lower'].apply(lambda x: tokenization(x))
data.head()
```

```python
#importing nlp library
import nltk
nltk.download('omw-1.4')
nltk.download('stopwords')
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
li = ['car','great','get','vehicle','’']
company_list = original['Company'].unique()
model_list=original['Model'].unique()
print(company_list)
print(model_list)
for x in company_list:
  stopwords.append(x)
for x in model_list:
  stopwords.append(x) 
for x in li:
  stopwords.append(x)
print(stopwords)
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords and not i.isnumeric()]
    return output
```

```python
data['no_stopwords']= data['msg_tokenied'].apply(lambda x:remove_stopwords(x))
```

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
```

```python
#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text
data['msg_stemmed']=data['no_stopwords'].apply(lambda x: stemming(x))
data.head()
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
#defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text
data['msg_lemmatized']=data['no_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
data.head()
```

```python
pd.set_option('display.max_colwidth',1)
#using only 2 columns
data= data [['Rating','msg_lemmatized']]
data.head()
```

```python
import gensim
import gensim.corpora as corpora
dict_ = corpora.Dictionary(data['msg_lemmatized'])
print(dict_)
```

```python
DTM = [dict_.doc2bow(i) for i in data['msg_lemmatized']]
DTM
```

```python
LDA = gensim.models.ldamodel.LdaModel
```

```python
LDAmodel = LDA(DTM, num_topics=11, id2word=dict_, passes=1, random_state=0, eval_every=None)
LDAmodel.print_topics()
```

```python
topics=['reliable','comfort','transmission','efficient','speed','mileage','safety','accessories','price','attractive','colour']
```

```python
count=0
aspect=[]
for i in LDAmodel[DTM]:
  print("doc : ",count,i)
  max=0
  pos=0
  for j in i:
    if max < j[1]:
      max = j[1]
      pos=j[0]
  print(pos)
  print("Favoured Topic...",topics[pos])
  aspect.append(topics[pos])
  count+=1
data=data.assign(Aspect=aspect)
```

```python
import pandas as pd
trainData = pd.read_csv("SVM TRAIN DATA.csv")
testData = pd.read_csv("SVM TEST DATA.csv")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Create feature vectors
vectorizer = TfidfVectorizer(min_df = 5,max_df = 0.8,sublinear_tf = True,use_idf = True)
train_vectors = vectorizer.fit_transform(trainData['Review'])
test_vectors = vectorizer.transform(testData['Review'])
```

```python
from sklearn import svm
from sklearn.metrics import classification_report
# Perform classification with SVM, kernel=linear
classifier_linear = svm.SVC(kernel='linear')
classifier_linear.fit(train_vectors, trainData['Rating'])
prediction_linear = classifier_linear.predict(test_vectors)

report = classification_report(testData['Rating'], prediction_linear, output_dict=True)

print('positive: ', report['pos'])
print('negative: ', report['neg'])
```

```python
review_vector=vectorizer.transform(original['Review'])
original['Label']=classifier_linear.predict(review_vector)
data['Label']=original['Label']
data['Review']=original['Review']
print(classifier_linear.predict(review_vector))
```
